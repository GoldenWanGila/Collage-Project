```python
import plotter
import pandas as pd
```

```python
dataFrame = pd.read_csv('inputs/railway.csv')
dataFrame_localTrain = dataFrame[dataFrame['num']>=500]
dataFrame_localTrain = dataFrame_localTrain[dataFrame_localTrain['date'] >= '2021/01/22']
dataFrame_localTrain
```

```python
dataFrame_localTrain_Fengyuan = dataFrame_localTrain[dataFrame_localTrain['Station']=='Fengyuan']
dataFrame_localTrain_Taiyuan = dataFrame_localTrain[dataFrame_localTrain['Station']=="Taiyuan"]
dataFrame_localTrain_Taichung = dataFrame_localTrain[dataFrame_localTrain['Station']=='Taichung']
dataFrame_localTrain_Xinwuri = dataFrame_localTrain[dataFrame_localTrain['Station']=='Xinwuri']
dataFrame_localTrain_Chenggong = dataFrame_localTrain[dataFrame_localTrain['Station']=='Tanzi']
dataFrame_localTrain_Changhua = dataFrame_localTrain[dataFrame_localTrain['Station']=='Changhua']
dataFrame_localTrain_Yuanlin = dataFrame_localTrain[dataFrame_localTrain['Station']=='Yuanlin']
```

```python
dataFrameList = [dataFrame_localTrain_Fengyuan, dataFrame_localTrain_Taiyuan, dataFrame_localTrain_Taichung, dataFrame_localTrain_Xinwuri, dataFrame_localTrain_Chenggong, dataFrame_localTrain_Changhua, dataFrame_localTrain_Yuanlin]
mseList = []
for positionStart in range(len(dataFrameList)):
    mseList.append([])
    dataFrameStart = dataFrameList[positionStart]
    for positionEnd in range(len(dataFrameList)):
        dataFrameEnd = dataFrameList[positionEnd]
        if positionStart > positionEnd:  # 順行
            x,y = plotter.arrangeDataFrame(dataFrameStart, dataFrameEnd, 0)
            mseList[positionStart].append(plotter.getRMSE(x,y))
        elif positionStart < positionEnd: # 逆行
            x,y = plotter.arrangeDataFrame(dataFrameStart, dataFrameEnd, 1)
            mseList[positionStart].append(plotter.getRMSE(y,x))
        else:
            mseList[positionStart].append('--')
```

```python
stationList = ['Fengyuan', 'Taiyuan', 'Taichung', 'Xinwuri', 'Chenggong', 'Changhua', 'Yuanlin']

mseDict = dict(zip(stationList, mseList))
mse_df = pd.DataFrame(mseDict, index=stationList)
```

```python
mse_df
```

```python
import datetime
```

```python
dataFrame_localTrain_weekday = dataFrame_localTrain[dataFrame_localTrain['date'].apply(lambda date: datetime.date(int(date[:4]), int(date[5:7]), int(date[8:])).isoweekday() < 6)]
dataFrame_localTrain_weekday
```

```python
dataFrame_localTrain_weekend = dataFrame_localTrain[dataFrame_localTrain['date'].apply(lambda date: datetime.date(int(date[:4]), int(date[5:7]), int(date[8:])).isoweekday() > 5)]
dataFrame_localTrain_weekend
```

## 平日尖峰

```python
# 上班
dataFrame_localTrain_weekday_onWork = dataFrame_localTrain_weekday[dataFrame_localTrain_weekday['time'] >= '06:00']
dataFrame_localTrain_weekday_onWork = dataFrame_localTrain_weekday_onWork[dataFrame_localTrain_weekday_onWork['time'] <= '08:00']
# 下班
dataFrame_localTrain_weekday_offWork = dataFrame_localTrain_weekday[dataFrame_localTrain_weekday['time'] >= '17:00']
dataFrame_localTrain_weekday_offWork = dataFrame_localTrain_weekday_offWork[dataFrame_localTrain_weekday_offWork['time'] <= '20:00']
```
